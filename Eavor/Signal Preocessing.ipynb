```python
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft,fftfreq
import numpy as np
from math import pi
%matplotlib inline
```

```python
#Generate sine wave
Fs=1600
t=np.arange(0,20,1/Fs)
f=20
```

```python
x=600*np.sin(2*pi*2*f*t)+400*np.sin(2*pi*5*f*t)+300*np.sin(2*pi*3*f*t)+450*np.sin(2*pi*8*f*t)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(2,1,1)
plt.plot(t,x)
```

```python
#compute FFTn
X=fft(x)
```

```python
#generate frequency axis
n=np.size(t)
fr=(Fs/2) * np.linspace(0,1,int(n/2))
X_m=(2/n)*np.abs(X[0:np.size(fr)])
```

```python
plt.figure(figsize=(10,8))
plt.plot(fr,X_m)
plt.xlabel("Frequency")
plt.ylabel('Amplitude')
plt.legend('Amplitude')
```

```python
# power spectrum density
pp=(2/n)*(np.abs(X[0:np.size(fr)])**2)
pp.size
```

```python
plt.figure(figsize=(10,8))
plt.plot(fr,pp)
plt.xlabel("Frequency")
plt.ylabel('Amplitude g2/Hz')
plt.legend('Amp')
```

```python
plt.figure(figsize=(10,8))
plt.xlabel("Time(s)")
plt.ylabel('Amplitude g2/Hz')
plt.legend('Amplitude')
plt.plot(t[0:800],pp)
```

```python
#Cross Checking power spectral values
mean_x=np.mean(x)
std_x=np.std(x)
var_x=std_x**2
print(mean_x,std_x,var_x)
```

```python
sample_rate = 1024
N = (2 - 0) * sample_rate
```

```python
time = np.linspace(0, 2, N)
```

```python
freq1 = 60
magnitude1 = 25
freq2 = 270
magnitude2 = 2

waveform1 = magnitude1 * np.sin (2 * pi * freq1 * time)
waveform2 = magnitude2 * np.sin (2 * pi * freq2 * time)
```

```python
noise = np.random.normal (0, 3, N)
```

```python
time_data = waveform1 + waveform2 + noise
```

```python
plt.plot (time [0:100], time_data [0:100])
plt.title ('Time Domain Signal')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
```

```python
frequency = np.linspace (0.0, 512, int (N/2))

freq_data = fft(time_data)
y = 2/N * np.abs (freq_data [0:np.int (N/2)])
```

```python
plt.plot(frequency, y)
plt.title('Frequency domain Signal')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
```
